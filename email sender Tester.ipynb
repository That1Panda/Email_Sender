```python
# Import modules
import smtplib, ssl
## email.mime subclasses
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
## The pandas library is only for generating the current date, which is not necessary for sending emails
import pandas as pd
from email.mime.application import MIMEApplication

# Define the HTML document
html = '''
    <html>
        <body>
            
            Dear Ex-Participant,<br><br>

            We hope this email finds you well.<br><br>
            
            
            <p> <b>STP's 18th year is about to begin,</b> over the past 17 years, 17 crews have exerted their efforts to deliver STP’s vision to take its role in developing our community. Every year a new team members gains fruitful <b>knowledge</b> that develops their mind and skills and <b>experience</b> new things through their work that develops them technically and non-technically in various fields. Every year a new team makes impressive <b>achievements</b> that leave them proud and meet new people and create unique <b>memories</b> that they never forget.</p>
            <p><b>STP '23 Members Recruitment are open ! </b>Now is your chance to write your own story.</p>
            
            
            
            <u>Application form:</u><br>
            https://www.stp-org.com/stp23/Member_Recruitment/Members_Form/<br>
            <u>Committees job description:</u><br>
            https://www.stp-org.com/stp23/Member_Recruitment/Event_Page/<br>
                
            <p>Also checkout our <a href="https://www.facebook.com/STP.Organization">facebook page</a> , you can find more details about the committees on our social media platforms. If you have any inquiries, don't hesitate to contact us through our facebook page.</p>
            <p>Sincerely,</p><br><br>
            --<br>
            Board Members | STP'23<br>
            <img src='cid:STP' width="100"><br>
            STP | Ain Shams University, Cairo, Egypt<br>
            https://www.stp-org.com/<br>
            --
        </body>
    </html>
    '''

def attach_file_to_email(email_message, filename, extra_headers=None):
    # Open the attachment file for reading in binary mode, and make it a MIMEApplication class
    with open(filename, "rb") as f:
        file_attachment = MIMEApplication(f.read())
    # Add header/name to the attachments    
    file_attachment.add_header(
        "Content-Disposition",
        f"attachment; filename= {filename}",
    )
    # Set up the input extra_headers for img
      ## Default is None: since for regular file attachments, it's not needed
      ## When given a value: the following code will run
         ### Used to set the cid for image
    if extra_headers is not None:
        for name, value in extra_headers.items():
            file_attachment.add_header(name, value)
    # Attach the file to the message
    email_message.attach(file_attachment)
    

# Set up the email addresses and password. Please replace below with your email address and password


# Generate today's date to be included in the email Subject
#date_str = pd.Timestamp.today().strftime('%Y-%m-%d')

# Create a MIMEMultipart class, and set up the From, To, Subject fields
email_message = MIMEMultipart()
email_message['From'] = email_sender
email_message['To'] = email_rec
email_message['Subject'] = "Members Recruitment | STP'23"

# Attach the html doc defined earlier, as a MIMEText html content type to the MIME message
email_message.attach(MIMEText(html, "html"))

attach_file_to_email(email_message, 'STP logo.png', {'Content-ID': '<STP>'})


# Convert it as a string
email_string = email_message.as_string()

# Connect to the Gmail SMTP server and Send Email
context = ssl.create_default_context()
with smtplib.SMTP_SSL("smtp.gmail.com", 465, context=context) as server:
    server.login(email_sender, password)
    server.sendmail(email_sender, email_rec, email_string)
```
